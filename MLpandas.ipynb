```python
# Pandas

import pandas as pd
```

```python
titanic_df = pd.read_csv('train.csv') # .read_확장자명('파일명')
titanic_df.head(4)
```

```python
# DataFrame의 행과 열 크기를 알아보는 방법
print(titanic_df.shape)
```

```python
titanic_df.info() #info() 메서드로 총 데이터 건수, 데이터 타입, Null 건수를 알 수 있다.
```

```python
titanic_df.describe() # describe() 메서드는 칼럼별 숫자형 데이터값의 분포도,평균값,최댓값,최솟값을 나타냄.  // 오직 숫자형(in,float 등)의 분포도만 조사하며 object 타입 자동 출력제외외
```

```python
# .counts 매서드 해당 칼럼의 유형과 건수, 데이터 타입 확인 가능
value_counts = titanic_df['Pclass'].value_counts()
print(value_counts)
```

```python
value_counts = titanic_df['Pclass'].value_counts()
print(type(value_counts))
print(value_counts)
```
